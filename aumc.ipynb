```python
import requests
from bs4 import BeautifulSoup

def search_uid(title):
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi"
    params = {
        "db": "pmc",
        "term": title,
        "retmode": "json"
    }
    response = requests.get(base_url, params=params)
    try:
        data = response.json()
        # 确保返回的ID是PMC ID
        uids = [id for id in data['esearchresult']['idlist'] if id.startswith('PMC')]
        return uids
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

def fetch_full_text(uid):
    base_url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi"
    params = {
        "db": "pmc",
        "id": uid,
        "retmode": "xml"
    }
    response = requests.get(base_url, params=params)
    return response.text

def extract_methodology(xml_text):
    soup = BeautifulSoup(xml_text, 'xml')
    sections = soup.find_all('sec')
    methodology_text = ''

    # 遍历所有节查找包含方法描述的部分
    for section in sections:
        title = section.find('title')
        if title and 'method' in title.text.lower():
            methodology_text = section.text.strip()
            break
    
    return methodology_text

# 使用论文标题进行搜索
title = "RASP: Optimal single puncta detection in complex cellular backgrounds"
uids = search_uid(title)
if uids:
    print(f"Found UIDs: {uids}")
    for uid in uids:
        xml_full_text = fetch_full_text(uid)  # 获取每个搜索结果的全文信息
        methodology = extract_methodology(xml_full_text)
        print("Methodology:", methodology)
else:
    print("No results found.")
    print(uids)
```

```python
import requests
import os

def download_from_github(repo, token):
    api_url = f'https://api.github.com/repos/{repo}/git/trees/master?recursive=1'
    headers = {'Authorization': f'token {token}'}
    
    response = requests.get(api_url, headers=headers)
    response.raise_for_status()
    
    tree = response.json().get('tree', [])
    txt_urls = [item['url'] for item in tree if item['path'].endswith('.txt') and item['type'] == 'blob']

    os.makedirs('github_downloads', exist_ok=True)

    for url in txt_urls:
        file_response = requests.get(url, headers=headers)
        file_response.raise_for_status()
        file_content = requests.get(file_response.json()['download_url']).content
        
        file_path = os.path.join('github_downloads', url.split('/')[-1])
        with open(file_path, 'wb') as file:
            file.write(file_content)
            print(f'Downloaded: {file_path}')

# 示例调用
download_from_github('IDR/idr-matadata', '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')
```

```python
import requests
import os

def download_from_github(repo, token, branch='master'):
    api_url = f'https://api.github.com/repos/{repo}/git/trees/{branch}?recursive=1'
    headers = {'Authorization': f'token {token}'}
    
    try:
        response = requests.get(api_url, headers=headers)
        response.raise_for_status()  # 检查请求是否成功
    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred: {err}")  # 输出HTTP错误信息
        return
    except Exception as err:
        print(f"An error occurred: {err}")  # 输出其他错误信息
        return

    tree = response.json().get('tree', [])
    txt_urls = [item['url'] for item in tree if item['path'].endswith('.txt') and item['type'] == 'blob']

    os.makedirs('github_downloads', exist_ok=True)

    for url in txt_urls:
        file_response = requests.get(url, headers=headers)
        file_response.raise_for_status()
        file_content = requests.get(file_response.json()['download_url']).content
        
        file_path = os.path.join('github_downloads', url.split('/')[-1])
        with open(file_path, 'wb') as file:
            file.write(file_content)
            print(f'Downloaded: {file_path}')

# 示例调用
download_from_github('IDR/idr-matadata', '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')  # 确保仓库名和分支名是正确的
```
